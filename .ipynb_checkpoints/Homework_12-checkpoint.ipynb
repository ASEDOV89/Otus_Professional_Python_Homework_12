```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set(style="whitegrid")

file_path = 'SampleSuperstore.csv'
data = pd.read_csv(file_path)

data.head()
# Просмотр первых нескольких строк
```

```python
data.info()
# Общая информация о датасете
```

```python
data.describe()
# Статистическая сводка числовых столбцов
```

```python
data.isnull().sum()
# Проверка на наличие пропущенных значений
```

```python
categorical_columns = ['Ship Mode', 'Segment', 'Country', 'City', 'State', 'Region', 'Category', 'Sub-Category']
for col in categorical_columns:
    print(f"{col}: {data[col].nunique()} unique values")
    print(data[col].value_counts().head())
    print('-' * 50)
#Анализ категорийных данных
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data['Sales'], kde=True, bins=30)
plt.title('Distribution of Sales')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
#Распределение продаж (Sales)
```

```python
top_cities_profit = data.groupby('City')['Profit'].sum().nlargest(10)
plt.figure(figsize=(10, 6))
sns.barplot(x=top_cities_profit.values, y=top_cities_profit.index, palette='viridis')
plt.title('Top 10 Cities by Profit')
plt.xlabel('Profit')
plt.ylabel('City')
plt.show()
#Топ-10 городов по прибыли (Profit)
```

```python
numeric_columns = ['Sales', 'Quantity', 'Discount', 'Profit']
correlation_matrix = data[numeric_columns].corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
#Корреляция между Sales, Quantity, Discount и Profit
```

```python
profit_by_category = data.groupby('Category')['Profit'].sum()
plt.figure(figsize=(8, 6))
sns.barplot(x=profit_by_category.index, y=profit_by_category.values, palette='Set3')
plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')
plt.show()
#Прибыль по категориям товаров
```

```python
#Сохраняем полученную статистику в отдельный файл по датасету
data.to_csv('sales_data.csv', index=False)
```
